```python
def countNa(df):
    """Pass your DataFrame, return NA count & % for each column """

    sum_null = df.isnull().sum()
    total = df.isnull().count()
    percent_nullvalues = 100* sum_null / total 
    df_null = pandas.DataFrame()
    df_null['Total'] = total
    df_null['Null_Count'] = sum_null
    df_null['Percent'] = round(percent_nullvalues,2)
    df_null = df_null.sort_values(by='Null_Count',ascending = False)
    df_null = df_null[df_null.Null_Count > 0]
    return df_null
```

```python
countNa(df)
```

```python
thr = 0.02

v_sup_na = int(df.shape[0]*thr/100)
print(v_sup_na)
df = df.dropna(thresh=v_sup_na, axis=1)
df.shape
```

```python
corr = df.corr()

def sortValues(df,col,nb_row_display):
    """ selec col for sort_values and choose number of row display, return  """
    z = df.sort_values(f"{col}", axis = 0, ascending = False).head(nb_row_display)
    return z  

corr_target = sortValues(corr,"SARS-Cov-2 exam result",40)
```
